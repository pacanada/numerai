```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

## the idea of this notebook is to explore the data provided by numeraire and some example of training and submission

```python
training_data = pd.read_csv("../numerai_datasets/numerai_training_data.csv")
# The tournament data is the data that Numerai uses to evaluate your model.
tournament_data = pd.read_csv("../numerai_datasets/numerai_tournament_data.csv")
```

```python
training_data = training_data.sample(1000)
tournament_data = tournament_data.sample(1000)
```

```python
training_cols = [c for c in training_data.columns if c.startswith("feature")]
target_col = "target"
```

```python
model = RandomForestRegressor(n_estimators=5)
```

```python
model.fit(X=training_data[training_cols], y=training_data[target_col] )
```

```python
model.predict(tournament_data[training_cols])
```
